```python
import pandas as pd

# low_memory=False заставит pandas прочитать весь файл целиком перед определением типов
df = pd.read_csv('dataset_2025-03-01_2026-03-29_external.csv', low_memory=False)

# Сразу отфильтруем "неопределенные" исходы (outcome_unknown == True)
df = df[df['outcome_unknown'] == False]

# Удалим битые воронки
df = df[df['lifecycle_incomplete'] == False]

print("Размер после базовой чистки:", df.shape)
```

```python
import pandas as pd
import numpy as np

def clean_data(df: pd.DataFrame) -> pd.DataFrame:
    """
    Функция для первичной очистки датасета и устранения Data Leakage.
    Оставляет только те признаки, которые строго известны в день оформления заказа (sale_ts).
    """
    # Создаем копию, чтобы избежать модификации исходного датафрейма (SettingWithCopyWarning)
    df_cleaned = df.copy()

    # ==========================================
    # 1. ФИЛЬТРАЦИЯ СТРОК (Удаление аномалий)
    # ==========================================

    # 1.1. Удаляем незавершенные сделки (исходы неизвестны)
    if 'outcome_unknown' in df_cleaned.columns:
        # Учитываем, что может быть как boolean True, так и строка 'TRUE'/'True'
        df_cleaned = df_cleaned[~df_cleaned['outcome_unknown'].astype(str).str.upper().isin(['TRUE', '1', '1.0'])]
        
    if 'lifecycle_incomplete' in df_cleaned.columns:
        df_cleaned = df_cleaned[~df_cleaned['lifecycle_incomplete'].astype(str).isin(['1', '1.0', 'True', 'TRUE'])]

    # 1.2. Пустые заказы (нет состава заказа)
    if 'lead_Состав заказа' in df_cleaned.columns:
        df_cleaned = df_cleaned.dropna(subset=['lead_Состав заказа'])
        df_cleaned = df_cleaned[df_cleaned['lead_Состав заказа'].astype(str).str.strip() != '']

    # 1.3. Предоплата (по условию оставляем только постоплату)
    if 'lead_Вид оплаты' in df_cleaned.columns:
        # Исключаем строки, где явно указана предоплата
        df_cleaned = df_cleaned[~df_cleaned['lead_Вид оплаты'].astype(str).str.lower().str.contains('предоплата', na=False)]

    # 1.4. Заказы менеджера-аномалии (~4000 сделок)
    if 'lead_responsible_user_id' in df_cleaned.columns:
        manager_counts = df_cleaned['lead_responsible_user_id'].value_counts()
        # Динамически находим менеджера с аномальным кол-вом сделок (например, от 3900 до 4100)
        anomaly_managers = manager_counts[(manager_counts > 3900) & (manager_counts < 4100)].index
        if len(anomaly_managers) > 0:
            df_cleaned = df_cleaned[~df_cleaned['lead_responsible_user_id'].isin(anomaly_managers)]

    # 1.5. Аномальные лаги доставки (112 - 147 дней)
    if 'days_to_outcome' in df_cleaned.columns:
        df_cleaned = df_cleaned[~df_cleaned['days_to_outcome'].between(112, 147)]
        
    # 1.6. Отрицательные дни доставки (сбои CRM)
    if 'days_handed_to_issued_pvz' in df_cleaned.columns:
        # Оставляем только те, где >= 0, либо где стоит NaN (если заказ еще не передан, но нам он нужен)
        df_cleaned = df_cleaned[(df_cleaned['days_handed_to_issued_pvz'] >= 0) | (df_cleaned['days_handed_to_issued_pvz'].isna())]

    # ==========================================
    # 2. УДАЛЕНИЕ МУСОРНЫХ КОЛОНОК И DATA LEAKAGE
    # ==========================================

    # Технические ID и трек-номера
    tech_cols = [
        'lead_id', 'contact_id', 'lead_account_id', 'lead_TRANID', 
        'lead__ym_uid', 'lead_yclid', 'lead_roistat', 'lead_FORMID',
        'lead_Номер отправления СДЭК', 'lead_Трек-номер'
    ]
    
    # Персональные и точные геоданные
    redacted_cols = [
        'contact_Телефон', 'contact_Email', 'contact_Адрес клиента', 'contact_Адрес ПВЗ'
    ]

    # Утечка данных (Data Leakage) - события после sale_ts
    leakage_cols = [
        'issued_or_pvz_ts', 'received_ts', 'rejected_ts', 'returned_ts', 'closed_ts', 
        'days_sale_to_handed', 'days_handed_to_issued_pvz', 'days_to_outcome', 
        'current_status_id', 'lead_status_id', 'lead_conditional_reject', 
        'lead_Дата создания накладной СДЭК', 'lead_Счет оплачен', 
        'lead_Оплачено клиентом', 'lead_Дата получения денег на Р/С', 'lead_Оплата МОП'
    ]

    # Собираем все колонки на удаление в один список
    all_cols_to_drop = tech_cols + redacted_cols + leakage_cols

    # Оставляем в списке на удаление только те колонки, которые реально присутствуют в df (чтобы избежать KeyError)
    cols_to_drop_existing = [col for col in all_cols_to_drop if col in df_cleaned.columns]
    
    # Удаляем колонки
    df_cleaned = df_cleaned.drop(columns=cols_to_drop_existing)

    # Дополнительная очистка: иногда [REDACTED] остаётся в строковых признаках.
    # Если нужно очистить вообще все колонки, где преобладает [REDACTED]:
    for col in df_cleaned.select_dtypes(include=['object', 'string']).columns:
        if df_cleaned[col].astype(str).str.contains(r'\[REDACTED\]').sum() / len(df_cleaned) > 0.9:
            df_cleaned = df_cleaned.drop(columns=[col])

    return df_cleaned
```

```python
import pandas as pd
import numpy as np
import re

# Если вдруг будут предупреждения о типах данных при чтении, можно добавить low_memory=False
df_raw = pd.read_csv('dataset_2025-03-01_2026-03-29_external.csv', low_memory=False)

print(f"🔥 Исходный датасет загружен.")
print(f"Строк: {df_raw.shape[0]}, Колонок: {df_raw.shape[1]}")
```

```python
# Применяем очистку
df_cleaned = clean_data(df_raw)

print(f"🧹 После первичной очистки (удаление аномалий и Data Leakage):")
print(f"Осталось строк: {df_cleaned.shape[0]} (удалили {df_raw.shape[0] - df_cleaned.shape[0]})")
print(f"Осталось колонок: {df_cleaned.shape[1]} (удалили {df_raw.shape[1] - df_cleaned.shape[1]})")
```

```python
def engineer_features(df: pd.DataFrame) -> pd.DataFrame:
    df_feat = df.copy()

    # ==========================================
    # 1. GARBAGE COLLECTION с выводом логов
    # ==========================================
    cols_before = df_feat.shape[1]
    
    # Удаляем колонки, где >80% NaN (оставляем те, где есть хотя бы 20% не-NaN)
    thresh_val = int(len(df_feat) * 0.2)
    df_feat = df_feat.dropna(axis=1, thresh=thresh_val)
    
    cols_after = df_feat.shape[1]
    dropped_cols = cols_before - cols_after
    
    print(f"🗑️ [Garbage Collection]:")
    print(f"   - Колонок до: {cols_before}")
    print(f"   - Удалено пустых колонок (>80% NaN): {dropped_cols}")
    print(f"   - Осталось колонок: {cols_after}\n")

    # ==========================================
    # 2. КАЛЕНДАРНЫЕ ПРИЗНАКИ
    # ==========================================
    if 'sale_ts' in df_feat.columns:
        sale_dt = pd.to_datetime(df_feat['sale_ts'], unit='s', errors='coerce')
        df_feat['sale_month'] = sale_dt.dt.month
        df_feat['sale_day'] = sale_dt.dt.day
        df_feat['sale_dayofweek'] = sale_dt.dt.dayofweek

    # ==========================================
    # 3. ПАРСИНГ ТЕКСТА: СОСТАВ ЗАКАЗА
    # ==========================================
    def parse_order_contents(text):
        res = {
            'total_items': 0, 'has_delivery_fee': False, 'total_goods_price': 0.0,
            'has_mask': 0, 'has_pillow': 0, 'has_bandage': 0
        }
        if not isinstance(text, str) or not text.strip():
            return pd.Series(res)

        blocks = re.split(r'(?:^|\n)(?=\d+\))', text.strip())
        for block in blocks:
            if not block.strip(): continue
            name_match = re.search(r'\d+\)\s*([^\n]+)', block)
            qty_match = re.search(r'Кол-во:\s*(\d+)', block)
            price_match = re.search(r'Розничная цена:\s*(\d+)', block)

            name = name_match.group(1).lower() if name_match else ""
            qty = int(qty_match.group(1)) if qty_match else 0
            price = float(price_match.group(1)) if price_match else 0.0

            if 'доставка' in name:
                if price > 0: res['has_delivery_fee'] = True
            else:
                res['total_items'] += qty
                res['total_goods_price'] += qty * price
                if 'маска' in name: res['has_mask'] = 1
                if 'подушка' in name: res['has_pillow'] = 1
                if 'повязка' in name: res['has_bandage'] = 1
        return pd.Series(res)

    if 'lead_Состав заказа' in df_feat.columns:
        print("⚙️ [Feature Engineering]: Парсинг 'lead_Состав заказа'...")
        parsed_df = df_feat['lead_Состав заказа'].apply(parse_order_contents)
        df_feat = pd.concat([df_feat, parsed_df], axis=1)
        # Удаляем исходную текстовую колонку, чтобы она не мешала ML-модели
        df_feat = df_feat.drop(columns=['lead_Состав заказа'])

    # ==========================================
    # 4. ОБРАБОТКА КАТЕГОРИАЛЬНЫХ ДАННЫХ
    # ==========================================
    if 'lead_Квалификация лида' in df_feat.columns:
        df_feat['lead_Квалификация лида'] = df_feat['lead_Квалификация лида'].fillna('Неизвестно')

    utm_cols = [col for col in df_feat.columns if 'utm' in col.lower()]
    for col in utm_cols:
        df_feat[col] = df_feat[col].fillna('organic')

    print("✅ Feature Engineering завершен.")
    return df_feat

# Применяем Feature Engineering
df_final = engineer_features(df_cleaned)

print(f"\n📊 Финальный размер датасета перед обучением:")
print(f"Строк: {df_final.shape[0]}")
print(f"Колонок (признаков): {df_final.shape[1]}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ML библиотеки
from catboost import CatBoostClassifier
from sklearn.metrics import (
    roc_auc_score, 
    precision_score, 
    recall_score, 
    f1_score, 
    confusion_matrix, 
    roc_curve
)

# Настройки для красивых графиков
sns.set_theme(style="whitegrid")
```

```python
print("⚙️ Подготовка данных для CatBoost...")

# 1. Подготовка таргета
# Удаляем строки с пустым таргетом (если такие случайно остались)
df_final = df_final.dropna(subset=['buyout_flag']).copy()

# Преобразуем таргет в int (1 / 0)
# Учитываем, что там могут быть строки 'True'/'False' или булевые значения
if df_final['buyout_flag'].dtype == 'object':
    df_final['buyout_flag'] = df_final['buyout_flag'].astype(str).str.lower().map({'true': 1, 'false': 0, '1': 1, '0': 0})
else:
    df_final['buyout_flag'] = df_final['buyout_flag'].astype(int)

# 2. Сортировка по времени (ОБЯЗАТЕЛЬНО ДЛЯ OUT-OF-TIME)
if 'sale_ts' in df_final.columns:
    df_final = df_final.sort_values(by='sale_ts', ascending=True).reset_index(drop=True)

# 3. Разделение признаков на числовые и категориальные
# Исключаем колонки, которые не пойдут в X
cols_to_exclude = ['buyout_flag', 'sale_ts', 'lead_price']
features = [col for col in df_final.columns if col not in cols_to_exclude]

# Определяем категориальные признаки (object, string, category)
cat_cols = df_final[features].select_dtypes(include=['object', 'category', 'string']).columns.tolist()
# Определяем числовые признаки
num_cols = df_final[features].select_dtypes(exclude=['object', 'category', 'string']).columns.tolist()

# 4. Заполнение пропусков
# Категориальные -> 'Unknown', приводим к строковому типу (требование CatBoost)
df_final[cat_cols] = df_final[cat_cols].fillna('Unknown').astype(str)

# Числовые -> 0
df_final[num_cols] = df_final[num_cols].fillna(0)

# Список категориальных фичей для передачи в модель
cat_features = cat_cols

print(f"Всего признаков для обучения: {len(features)}")
print(f"Категориальных признаков: {len(cat_features)}")
```

```python
print("✂️ Разбиение на Train (80%) и Test (20%) по времени...")

# Вычисляем индекс для разделения (80% данных)
split_idx = int(len(df_final) * 0.8)

# Разбиваем датафреймы (shuffle=False происходит автоматически при iloc)
train_df = df_final.iloc[:split_idx]
test_df = df_final.iloc[split_idx:]

# Формируем X и y
X_train = train_df[features]
y_train = train_df['buyout_flag']

X_test = test_df[features]
y_test = test_df['buyout_flag']

print(f"Размер Train: X={X_train.shape}, y={y_train.shape}")
print(f"Размер Test:  X={X_test.shape}, y={y_test.shape}")
```

```python
print("🚀 Запуск обучения CatBoost...")

# Инициализируем модель
cb_model = CatBoostClassifier(
    iterations=1000,
    eval_metric='AUC',
    random_seed=42,
    early_stopping_rounds=50,
    # Если на машине/в колабе есть GPU, раскомментируй строку ниже:
    # task_type='GPU', 
    verbose=100  # Выводить лог каждые 100 итераций
)

# Обучаем модель
cb_model.fit(
    X_train, y_train,
    cat_features=cat_features,
    eval_set=(X_test, y_test),
    plot=False # В Jupyter можно поставить True, чтобы рисовать график прямо во время обучения
)

print("✅ Обучение завершено!")
```

```python
# Получаем предсказания классов и вероятностей
y_pred = cb_model.predict(X_test)
y_pred_proba = cb_model.predict_proba(X_test)[:, 1]

# Рассчитываем метрики
auc = roc_auc_score(y_test, y_pred_proba)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("📊 Метрики на тестовой (Out-of-Time) выборке:")
print(f"ROC-AUC:   {auc:.4f}")
print(f"Precision: {precision:.4f} (насколько точны наши предсказания 'Выкупа')")
print(f"Recall:    {recall:.4f} (какую долю реальных выкупов мы нашли)")
print(f"F1-score:  {f1:.4f}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# 1. Матрица ошибок (Confusion Matrix)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[0], 
            cbar=False, annot_kws={"size": 14})
axes[0].set_title('Confusion Matrix\n(0 = Отказ, 1 = Выкуп)', fontsize=14, pad=15)
axes[0].set_xlabel('Предсказанный класс', fontsize=12)
axes[0].set_ylabel('Истинный класс', fontsize=12)

# 2. ROC-кривая (ROC-Curve)
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
axes[1].plot(fpr, tpr, color='#ff7f0e', lw=2, label=f'CatBoost (AUC = {auc:.3f})')
axes[1].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', alpha=0.6)
axes[1].set_title('ROC Curve', fontsize=14, pad=15)
axes[1].set_xlabel('False Positive Rate (FPR)', fontsize=12)
axes[1].set_ylabel('True Positive Rate (TPR)', fontsize=12)
axes[1].legend(loc="lower right", fontsize=12)
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Настройки для графиков
sns.set_theme(style="whitegrid")

print("🔍 Анализ базовой важности признаков CatBoost...")

# Извлекаем важность признаков
feature_importances = cb_model.get_feature_importance()
importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': feature_importances
}).sort_values(by='Importance', ascending=False)

# Берем Топ-15
top_15_features = importance_df.head(15)

# Строим график
plt.figure(figsize=(10, 8))
sns.barplot(
    x='Importance', 
    y='Feature', 
    data=top_15_features, 
    palette='viridis'
)
plt.title('Топ-15 признаков по встроенной важности (CatBoost)', fontsize=14, pad=15)
plt.xlabel('Важность', fontsize=12)
plt.ylabel('Признак', fontsize=12)
plt.tight_layout()
plt.show()

# Выводим текстом для быстрого анализа
print("\nТоп-5 кандидатов на проверку утечки данных (Data Leakage):")
for idx, row in top_15_features.head(5).iterrows():
    print(f"- {row['Feature']}: {row['Importance']:.2f}%")
```

```python
import shap

print("🧠 Расчет SHAP-значений (это может занять пару минут)...")

# Инициализируем TreeExplainer для нашей модели
explainer = shap.TreeExplainer(cb_model)

# Считаем SHAP для тестовой выборки
shap_values = explainer.shap_values(X_test)

# Строим Summary Plot
plt.figure(figsize=(12, 8))
plt.title('SHAP Summary Plot (Влияние признаков на предсказание)', fontsize=14)
# plot_type="dot" показывает плотность и направление влияния
shap.summary_plot(shap_values, X_test, max_display=15, show=False)
plt.tight_layout()
plt.show()
```

```python
print("💰 Запуск финансовой симуляции...")

LOGISTICS_COST = 1000 # Стоимость холостой доставки (туда-обратно)
thresholds = np.arange(0.1, 0.95, 0.05)

# Проверяем, что индексы в X_test, y_test и test_df совпадают
# lead_price берем из test_df (мы его туда не передавали в модель, но в датафрейме он остался)
lead_prices = test_df['lead_price'].values
actual_outcomes = y_test.values
pred_probas = cb_model.predict_proba(X_test)[:, 1]

profits = []

# Базовая линия: отправляем ВСЕ заказы без разбора
# Зарабатываем на тех, кто выкупил, теряем на тех, кто отказался
baseline_profit = np.sum(lead_prices[actual_outcomes == 1]) - (np.sum(actual_outcomes == 0) * LOGISTICS_COST)

for thresh in thresholds:
    # Модель говорит "Выкупит", если вероятность >= порога
    preds = (pred_probas >= thresh).astype(int)
    
    # Считаем финансовый результат
    # 1. True Positives (Модель: 1, Факт: 1) -> Получаем прибыль = lead_price
    tp_mask = (preds == 1) & (actual_outcomes == 1)
    profit_tp = np.sum(lead_prices[tp_mask])
    
    # 2. False Positives (Модель: 1, Факт: 0) -> Теряем на доставке = LOGISTICS_COST
    fp_mask = (preds == 1) & (actual_outcomes == 0)
    loss_fp = np.sum(fp_mask) * LOGISTICS_COST
    
    # 3. Predict 0 -> Не отправляем, прибыль 0, убыток 0
    # Итоговая прибыль:
    total_profit = profit_tp - loss_fp
    profits.append(total_profit)

# Ищем оптимальный порог
best_idx = np.argmax(profits)
best_threshold = thresholds[best_idx]
max_profit = profits[best_idx]

# Строим график
plt.figure(figsize=(12, 6))
plt.plot(thresholds, profits, marker='o', lw=2, color='#2ca02c', label='Прибыль с ML-моделью')
plt.axhline(y=baseline_profit, color='#d62728', linestyle='--', lw=2, label='Baseline (отправляем всем)')
plt.axvline(x=best_threshold, color='blue', linestyle=':', lw=2, label=f'Опт. порог: {best_threshold:.2f}')

# Наводим красоту
plt.title('Зависимость бизнес-прибыли от порога уверенности ML-модели', fontsize=15, pad=15)
plt.xlabel('Порог вероятности (Threshold)', fontsize=12)
plt.ylabel('Итоговая прибыль (Руб.)', fontsize=12)
plt.xticks(thresholds)
# Форматируем ось Y в миллионы/тысячи для читаемости
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x))))
plt.legend(fontsize=12)
plt.grid(True, alpha=0.4)
plt.tight_layout()
plt.show()

# Выводим бизнес-выводы
print("📊 ФИНАНСОВЫЕ ИТОГИ СИМУЛЯЦИИ:")
print(f"Прибыль без ML (отправляем все): {int(baseline_profit):,} руб.")
print(f"Максимальная прибыль с ML:       {int(max_profit):,} руб.")
print(f"Разница (Бизнес-эффект):         +{int(max_profit - baseline_profit):,} руб.")
print(f"Оптимальный порог отсечения:     {best_threshold:.2f}")

# Анализ метрики
if (max_profit - baseline_profit) > baseline_profit * 2:
    print("\n⚠️ ВНИМАНИЕ: Слишком гигантский бизнес-эффект. Это подтверждает теорию о Data Leakage.")
    print("Внимательно изучи верхние фичи на графике SHAP. Возможно, модель опирается на статусы отмены из CRM.")
```
